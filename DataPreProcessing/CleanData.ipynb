```python
import pandas as pd

df = pd.read_csv('../hotel.csv') 
```

+ Booking_ID : mã định danh duy nhất của mỗi lần đặt phòng
+ no_of_adults : Số lượng người lớn
+ no_of_children : Số lượng trẻ em
+ no_of_weekend_nights : Số đêm cuối tuần (thứ bảy hoặc chủ nhật) mà khách đã ở hoặc đặt phòng tại khách sạn
+ no_of_week_nights : Số đêm trong tuần (từ thứ Hai đến thứ Sáu) mà khách đã ở hoặc đặt phòng tại khách sạn
+ type_of_meal_plan : Loại gói bữa ăn mà khách hàng đặt
+ required_car_parking_space : Khách hàng có yêu cầu chỗ đậu xe không? (0 - Không, 1- Có)
+ room_type_reserved : Loại phòng được khách hàng đặt
+ lead_time : Số ngày giữa ngày đặt phòng và ngày đến
+ arrival_year : Năm đến
+ arrival_month : Tháng đến
+ arrival_date : Ngày trong tháng
+ market_segment_type : Chỉ định phân khúc thị trường
+ repeat_guest : Khách hàng có phải là khách quen không? (0 - Không, 1- Có)
+ no_of_previous_cancellations : Số lượng đặt phòng trước đó đã bị khách hàng hủy trước khi đặt phòng hiện tại
+ no_of_previous_bookings_not_canceled : Số lượng đặt phòng trước đó không bị khách hàng hủy trước khi đặt phòng hiện tại
+ avg_price_per_room : Giá trung bình mỗi ngày của đặt phòng. (tính bằng euro)
+ no_of_special_requests : Tổng số yêu cầu đặc biệt của khách hàng (ví dụ: tầng cao, tầm nhìn từ phòng, v.v.)
+ booking_status : Cho biết đặt phòng đã bị hủy hay chưa. (1 - Not_Canceled, 0 - Canceled)

Kiểm tra các giá trị trong các cột có kiểu dữ liệu object

```python
col = 'type_of_meal_plan'
print(f"Cột: {col}")
print("Số lượng giá trị unique:", df[col].nunique())
print(df[col].value_counts().to_string())
```

```python
col = 'room_type_reserved'
print(f"Cột: {col}")
print("Số lượng giá trị unique:", df[col].nunique())
print(df[col].value_counts().to_string())
```

```python
col = 'market_segment_type'
print(f"Cột: {col}")
print("Số lượng giá trị unique:", df[col].nunique())
print(df[col].value_counts().to_string())
```

```python
col = 'booking_status'
print(f"Cột: {col}")
print("Số lượng giá trị unique:", df[col].nunique())
print(df[col].value_counts().to_string())
```

Encoding các cột kiểu dữ liệu object

```python
from sklearn.preprocessing import LabelEncoder

cols_to_encode = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type', 'booking_status']
label_encoders = {}

for col in cols_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

Lưu file sau khi đã làm sạch dữ liệu

```python
df.to_csv('../clean.csv', index=False, encoding='utf-8')
```
