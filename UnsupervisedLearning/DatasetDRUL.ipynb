DATASET BAN ĐẦU -> GIẢM CHIỀU -> UNSUPERVISED LEARNING

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture

df = pd.read_csv('../DatasetDR.csv') 
```

```python
df.head()
```

```python
X = df.drop(columns=['booking_status'])
y = df['booking_status']
```

DBSCAN để phát hiện outliers và phân cụm

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

* Thêm vào DataFrame

```python
df['dbscan_cluster'] = dbscan_labels
df['dbscan_outlier'] = (dbscan_labels == -1).astype(int)
```

GMM để lấy xác suất thuộc các cụm

```python
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)
gmm_probs = gmm.predict_proba(X)
```

* Thêm vào DataFrame

```python
for i in range(gmm_probs.shape[1]):
    df[f'gmm_prob_{i}'] = gmm_probs[:, i]
```

Lưu dữ liệu

```python
df.to_csv('../DatasetDRUL.csv', index=False)
```
